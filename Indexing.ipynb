```python
def return_data():
    return_values = ['Benin', 'Angola', 'Algeria', 'Peru']
    continent_values = ['Africa', 'South America']
    values = df.query("Continent in @continent_values and Country in @return_values")
    return values
return_data()
```

```python
def return_data():
    return_values = ['Benin', 'Angola', 'Algeria', 'Peru']
    continent_values = ['Africa', 'North America']
    values = df.loc[['Continent'].str.contains(continent_values) & df.index.str.contains(return_values)]
    return values

return_data()
```
